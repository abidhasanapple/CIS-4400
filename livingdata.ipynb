```python
import json
import pandas as pd
```

```python
json_file_path = "New York_Kings County.json"
with open(json_file_path) as file:
    data = json.load(file)

occupational_areas = ["Management", "Computer & Mathematical", "Healthcare Practitioners & Technical"]
filtered_data = [d for d in data["annualSalaryList"] if d["occupationalArea"] in occupational_areas]

# create a dataframe based on json
king_df = pd.DataFrame(filtered_data)
king_df["county"] = "Brooklyn"
king_df_wide = pd.pivot(king_df, index='county', columns='occupationalArea', values='typicalAnnualSalary')
king_df_wide
```

```python
# Concatenate the dataframes vertically (row-wise)
ny_df = pd.concat([king_df, bronx_df, queens_df, richmond_df,newyork_df], axis=0)

# Reset the index of the merged dataframe
ny_df = ny_df.reset_index(drop=True)

ny_df_wide=pd.pivot(ny_df, index=['county'], columns = 'occupationalArea',values = 'typicalAnnualSalary') #Reshape from long to wide
ny_df_wide
```
